### What kinds of improvements are appropriate today?

A good Day 29 improvement is:

- Small enough to complete during class
- Clear enough that the repository owner can review it quickly
- Useful to a future reader or collaborator

Examples include:

- Improving the `README.md`
- Adding or revising rerun instructions

- Cleaning up notebook section headers

Avoid large rewrites, major feature additions, or changes that would require redesigning the entire project.

```bash
# Clone the repository
# Replace USERNAME and REPO-NAME with the correct values
git clone https://github.com/USERNAME/REPO-NAME.git

# Move into the repository directory
cd REPO-NAME

# Create and switch to a new branch
# Use a descriptive branch name, not something generic like 'new-branch' (see note below)
git checkout -b peer-review-fix-readme

# Confirm your current branch
git branch

### Branch naming note

Your branch name should describe the change clearly. Good examples:

- `peer-review-fix-readme`
- `improve-rerun-instructions`
- `clarify-model-limitations`
- `reorganize-notebook-headings`

Avoid vague names like:

- `branch1`
- `newstuff`
- `test`
- `day29`

### Checkpoint

After cloning and branching, answer the questions below. **If you have trouble with any of these steps, ask for help before proceeding!**


> **Local repo folder name:**
>
> **Branch name you created:**
>

- - -

<a id="plan"></a>

## 4. Plan the improvement

Before you start editing files, be explicit about what you are changing.

### Improvement planning checklist

Mark these mentally as you work through your change:

- Is the change scoped small enough for one class period?
- Will the improvement make the repo easier to understand, rerun, or review?
- Does the change preserve the original author's work rather than replacing it unnecessarily?
- Can you explain the value of the improvement in one or two sentences?

### Write your plan

Complete the prompts below before moving on.


> **File(s) you expect to edit:**
>
> **Type of improvement (documentation / notebook organization / interpretation / reproducibility / other):**
>
> **What problem are you trying to solve for the repo owner or future reader?**
>
> **How will you know when the improvement is complete?**


- - -

<a id="implement"></a>

## 5. Implement the improvement

Now make the change in the repository.

### While you work, keep these expectations in mind

- Keep the improvement focused and readable
- Prefer clear, specific writing over long or vague writing
- If you change a notebook, make sure the structure still makes sense to a reader
- If you change code, verify that the change does not break the workflow
- Do not delete important work just to simplify the repository

### Common improvement categories

#### A. README improvement

Possible updates:
- Clearer project summary
- Better rerun instructions
- Better explanation of data sources or S3 paths
- Better explanation of the modeling or SQL workflow
- Better statement of assumptions, limitations, or next steps

#### B. Notebook or report improvement

Possible updates:
- Add clearer section headers
- Add short interpretive notes after outputs
- Clarify what a metric means
- Improve markdown explanations between code cells

#### C. Reproducibility improvement

Possible updates:
- Add or improve step-by-step rerun instructions
- Clarify expected file locations or folder structure
- Explain what someone must already have configured before rerunning the work


### Work log

During or after implementing your change, briefly record what you actually edited.


> *List the file(s) you edited and what you changed:*



### Quick self-check before commit

Answer these questions before you move on:

- Is the change understandable without extra explanation from you?
- Did you improve something real rather than just rewording for style?
- Would you be comfortable having your own repo reviewed this way?
- If relevant, did you confirm that the notebook or code still runs?


- - -

<a id="pr"></a>

## 6. Commit, push, and open a pull request

Once your improvement is complete, commit it with a meaningful message and push your branch to GitHub.

### Suggested Git commands

Run these commands from inside the cloned repository.


```bash
# Check what changed
git status

# Stage your updated files (make sure you are adding the right ones!)
git add .

# Commit with a clear message (make sure it aligns with the improvement you made)
git commit -m "Improve README instructions for project rerun"

# Push your branch to GitHub (make sure to replace the branch name with your actual branch name)
git push -u origin peer-review-fix-readme
```

### Pull request checklist

After pushing your branch, open a pull request on GitHub. For details about creating a pull request, see the [GitHub documentation](https://docs.github.com/en/pull-requests/collaborating-with-pull-requests/proposing-changes-to-your-work-with-pull-requests/creating-a-pull-request).

You should be able to go to the "Pull requests" tab of the repository on GitHub and see a button to create a "New pull request" Or, alternatively, you can navigate to the branch you just created from the main page of the GitHub repository and it should prompt you to open a pull request for your recently pushed branch.

Your pull request should include:

- A clear title
- A short summary of what changed
- A short explanation of why the change was needed
- A note telling the reviewer what to look at first

### Example pull request description

You may adapt something like this:

> I updated the README to make the project workflow easier to rerun. The main change was adding clearer setup steps and a short note about assumptions and limitations. Please review the README first, especially the rerun section and the limitations paragraph.

## 7. Review an incoming pull request

Now switch roles. Someone should submit a pull request to **your** repository, and you need to review it.

The goal is not to automatically approve the PR. Instead, decide whether the change is actually ready to merge.

### Review options

Choose one of the following outcomes:

- **Accept as is**: the change is clear, useful, and ready to merge
- **Request comments or clarification**: the improvement seems useful, but the intent or details are not fully clear
- **Request additional work**: the idea is promising, but something important still needs to be revised before merge

**Note:** Regardless of your decision, you should leave a comment for the contributor indicating your decision and explaining your reasoning. If you request changes, be specific about what needs to be revised and why.

### What to look for

- Does the PR actually improve the repository?
- Is the change easy to understand?
- Is anything misleading, incomplete, or inaccurate?
- Does the change match the repo's existing structure and purpose?
- Would you merge this into your project right now?

### Optional GitHub review comment starter

If you are not sure how to phrase your feedback, adapt one of these:

- **Accept as is**: "Thanks — this is a clear and useful improvement. I reviewed the updated file(s), and I am comfortable merging this as is."
- **Request clarification**: "Thanks — this looks helpful overall. Before I merge it, can you clarify [specific point] so I understand the reasoning behind the change?"
- **Request additional work**: "Thanks for the contribution. I think the direction is good, but before I merge this, I need [specific revision] so the repository stays clear and accurate.


### Finalizing the pull request

Once you are satisfied with the change (which may require waiting for the author to make revisions), you can merge the pull request on GitHub. After merging, **make sure to pull the latest changes to your local repository so that your local `main` branch is up to date!**
