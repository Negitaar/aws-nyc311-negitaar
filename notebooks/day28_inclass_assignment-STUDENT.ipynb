## 3. Track the SageMaker workflow

As you run the provided notebook, pay attention to the overall sequence. According to the course guide, the SageMaker built-in workflow follows this general pattern:

1. Prepare the data
2. Convert it to the required all-numeric, label-first CSV format
3. Upload train/test files to S3
4. Configure a SageMaker `Estimator`
5. Run `.fit()` to launch a managed training job
6. Use Batch Transform to generate predictions
7. Bring predictions back into the notebook and compute metrics

This is the key conceptual difference from sklearn: your notebook is not doing all of the training itself. Instead, it is **describing and launching** a separate managed training job.
